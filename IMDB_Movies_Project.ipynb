```python
# - avg runtime , highest runtime movies, lowest runtime movies
avergeruntime = df['Runtime (Minutes)'].sum()
avergeruntime = avergeruntime/len(df['Runtime (Minutes)'])
print('The Average runtime :' , avergeruntime)

Highestruntime = df['Runtime (Minutes)'].max()
print('The Highest runtime :' , Highestruntime)

leastruntime = df['Runtime (Minutes)'].min()
print('The least runtime :' , leastruntime)


# This shows the average time in mins
```

```python
# Lets check if there is any relation between duration and revenue

def rating(value):
    if (value >= 9.0):
        return 'Blockbuster'
    elif (value < 9.0) and (value >= 7.0):
        return 'Super Hit'
    elif (value < 7.0) and (value >= 6.0):
        return 'Hit'
    elif (value < 6.0) and (value >= 5.0):
        return 'Average'
    elif (value < 5.0) and (value >= 0.1):
        return 'Flop'
    else:
        return 'NA'
```

```python
df['Movie Status'] = df['Rating'].apply(rating)
```

```python
df.sample(5)
```

```python
highruntimeduration = df['Runtime (Minutes)'] > 114.0
leastruntimeduration = df['Runtime (Minutes)'] <= 114.0
```

```python
highruntimeduration = df[highruntimeduration]
leastruntimeduration = df[leastruntimeduration]
```

```python
fig , ax = plt.subplots(figsize= (8, 6))
ax = sns.countplot(highruntimeduration['Movie Status'])
plt.title('Movie status of the movie with duration higher than average runtime')


for value in ax.patches:
    ax.annotate(f'{value.get_height()}', (value.get_x()+ value.get_width()/ 2., value.get_height()), ha = 'center', va= 'bottom')
    
    
plt.show()
```

```python
fig , ax = plt.subplots(figsize= (8, 6))
ax = sns.countplot(leastruntimeduration['Movie Status'])
plt.title('Movie status of the movie with duration lower than average runtime')

for values in ax.patches:
    ax.annotate(f'{values.get_height()}', (values.get_x()+values.get_width()/2., values.get_height()), ha = 'center' , va = 'bottom')
plt.show()
```
